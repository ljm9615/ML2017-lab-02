```python
# write your code here
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import random


def get_data(path):
    data = load_svmlight_file(path)
    return data[0].toarray(), data[1]


def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))


def get_loss(X, y, w, b):
    loss = 0
    for i in range(X.shape[0]):
        y_ = sigmoid(np.dot(X[i].data, w) + b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        loss += max(0, 1 - y[i] * y_)
    return loss / X.shape[0] + np.sum(w ** 2) / 2


def compute_gradient(X, y, w, b, sample_indices):
    # loss_function wb
    G_w = 0
    G_b = 0
    for i in sample_indices:
        y_ = sigmoid(np.dot(X[i].data, w) + b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        if 1 - y[i] * y_ >= 0:
            G_w += y[i] * X[i].data * (-1)
            G_b += y[i] + (-1)
        else:
            G_w += 0
            G_b += 0
    G_w = G_w / len(sample_indices) + w
    G_b = G_b / len(sample_indices)

    # 
    w = w - learning_rate * G_w
    b = b - learning_rate * G_b
    return w, b


def NAG(X, y, w, b, v_, learning_rate, sample_indices):
    v_w = v_[:-1]
    v_b = v_[-1:]
    Gamma = 0.9
    # loss_function wb
    G_w = 0
    G_b = 0
    for i in sample_indices:
        y_ = sigmoid(np.dot(X[i].data, w - Gamma * v_w) + b - Gamma * v_b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        if 1 - y[i] * y_ >= 0:
            G_w += y[i] * X[i].data * (-1)
            G_b += y[i] + (-1)
        else:
            G_w += 0
            G_b += 0
    G_w = G_w / len(sample_indices) + w
    G_b = G_b / len(sample_indices)

    # 
    v_w = Gamma * v_w + learning_rate * G_w
    v_b = Gamma * v_b + learning_rate * G_b
    w = w - v_w
    b = b - v_b

    return w, b, np.concatenate([v_w, v_b])


def RMSProp(X, y, w, b, Gt, learning_rate, sample_indices):
    # loss_function wb
    G_w = 0
    G_b = 0
    Gt_w = Gt[:-1]
    Gt_b = Gt[-1:]
    Gamma = 0.9
    epsilon = np.exp(-8)
    for i in sample_indices:
        y_ = sigmoid(np.dot(X[i].data, w) + b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        if 1 - y[i] * y_ >= 0:
            G_w += y[i] * X[i].data * (-1)
            G_b += y[i] + (-1)
        else:
            G_w += 0
            G_b += 0
    G_w = G_w / len(sample_indices) + w
    G_b = G_b / len(sample_indices)

    # 
    Gt_w = Gt_w * Gamma + (1 - Gamma) * np.multiply(G_w, G_w)
    Gt_b = Gt_b * Gamma + (1 - Gamma) * np.multiply(G_b, G_b)

    w = w - np.multiply(learning_rate * pow(Gt_w + epsilon, -1/2), G_w)
    b = b - np.multiply(learning_rate * pow(Gt_b + epsilon, -1/2), G_b)

    return w, b, np.concatenate([Gt_w, Gt_b])


def AdaDelta(X, y, w, b, Gt, delta, sample_indices):
    # loss_function wb
    G_w = 0
    G_b = 0
    Gt_w = Gt[:-1]
    Gt_b = Gt[-1:]
    delta_w = delta[:-1]
    delta_b = delta[-1:]
    Gamma = 0.95
    epsilon = pow(10, -8)
    for i in sample_indices:
        y_ = sigmoid(np.dot(X[i].data, w) + b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        if 1 - y[i] * y_ >= 0:
            G_w += y[i] * X[i].data * (-1)
            G_b += y[i] + (-1)
        else:
            G_w += 0
            G_b += 0
    G_w = G_w / len(sample_indices) + w
    G_b = G_b / len(sample_indices)

    # 
    Gt_w = Gt_w * Gamma + (1 - Gamma) * np.multiply(G_w, G_w)
    Gt_b = Gt_b * Gamma + (1 - Gamma) * np.multiply(G_b, G_b)
    step_w = np.multiply((-1) * pow(delta_w + epsilon, 1/2) * pow(Gt_w + epsilon, -1/2), G_w)
    step_b = np.multiply((-1) * pow(delta_b + epsilon, 1 / 2) * pow(Gt_b + epsilon, -1 / 2), G_b)
    w = w + step_w
    b = b + step_b
    delta_w = Gamma * delta_w + np.multiply((1 - Gamma) * step_w, step_w)
    delta_b = Gamma * delta_b + np.multiply((1 - Gamma) * step_b, step_b)

    return w, b, np.concatenate([Gt_w, Gt_b]), np.concatenate([delta_w, delta_b])


def Adam(X, y, w, b, Gt, mt, sample_indices, learning_rate, t):
    # loss_function wb
    G_w = 0
    G_b = 0
    Gt_w = Gt[:-1]
    Gt_b = Gt[-1:]
    mt_w = mt[:-1]
    mt_b = mt[-1:]
    Beta = 0.9
    Gamma = 0.999
    epsilon = pow(10, -8)
    for i in sample_indices:
        y_ = sigmoid(np.dot(X[i].data, w) + b)
        if y_ > 0.5:
            y_ = 1
        else:
            y_ = -1
        if 1 - y[i] * y_ >= 0:
            G_w += y[i] * X[i].data * (-1)
            G_b += y[i] + (-1)
        else:
            G_w += 0
            G_b += 0
    G_w = G_w / len(sample_indices) + w
    G_b = G_b / len(sample_indices)

    # 
    mt_w = mt_w * Beta + (1 - Beta) * G_w
    mt_b = mt_b * Beta + (1 - Beta) * G_b
    Gt_w = Gt_w * Gamma + (1 - Gamma) * np.multiply(G_w, G_w)
    Gt_b = Gt_b * Gamma + (1 - Gamma) * np.multiply(G_b, G_b)
    alpha = learning_rate * pow(1 - pow(Gamma, t), 1/2) * pow(1 - pow(Beta, t), -1/2)
    w = w - alpha * mt_w * pow(Gt_w + epsilon, -1/2)
    b = b - alpha * mt_b * pow(Gt_b + epsilon, -1/2)
    return w, b, np.concatenate([Gt_w, Gt_b]), np.concatenate([mt_w, mt_b])


def get_NAG(epoch, learning_rate):
    # 
    w = np.random.normal(size=123)
    b = np.random.normal(size=1)
    v = np.zeros(124)

    # loss
    val_losses = []

    for num in range(epoch):
        # SGD10
        sample_indices = random.sample(range(X_test.shape[0]), 100)

        # loss
        val_loss = get_loss(X_test, y_test, w, b)

        w, b, v = NAG(X_test, y_test, w, b, v, learning_rate, sample_indices)

        # loss
        val_losses.append(val_loss)
        print("NAG: ", num, val_loss)
    return val_losses


def get_RMSProp(epoch, learning_rate):
    # 
    w = np.random.normal(size=123)
    b = np.random.normal(size=1)
    Gt = np.zeros(124)

    # loss
    val_losses = []

    for num in range(epoch):
        # SGD10
        sample_indices = random.sample(range(X_test.shape[0]), 100)

        # loss
        val_loss = get_loss(X_test, y_test, w, b)

        w, b, Gt = RMSProp(X_test, y_test, w, b, Gt, learning_rate, sample_indices)

        # loss
        val_losses.append(val_loss)
        print("RMSProp: ", num, val_loss)
    return val_losses


def get_AdaDelta(epoch):
    # 
    w = np.random.normal(size=123)
    b = np.random.normal(size=1)
    Gt = np.zeros(124)
    delta = np.zeros(124)

    # loss
    val_losses = []

    for num in range(epoch):
        # SGD10
        sample_indices = random.sample(range(X_test.shape[0]), 100)

        # loss
        val_loss = get_loss(X_test, y_test, w, b)

        w, b, Gt, delta = AdaDelta(X_test, y_test, w, b, Gt, delta, sample_indices)

        # loss
        val_losses.append(val_loss)
        print("AdaDelta: ", num, val_loss)
    return val_losses


def get_Adam(epoch, learning_rate):
    # 
    w = np.random.normal(size=123)
    b = np.random.normal(size=1)
    Gt = np.zeros(124)
    mt = np.zeros(124)

    # loss
    val_losses = []

    for num in range(epoch):

        # SGD10
        sample_indices = random.sample(range(X_test.shape[0]), 100)
        # loss
        val_loss = get_loss(X_test, y_test, w, b)

        w, b, Gt, mt = Adam(X_test, y_test, w, b, Gt, mt, sample_indices, learning_rate, num + 1)

        # loss
        val_losses.append(val_loss)
        print("Adam: ", num, val_loss)
    return val_losses


if __name__ == "__main__":

    # 
    X_train, y_train = get_data(r"E:\machine learning\lab2\a9a")
    X_test, y_test = get_data(r"E:\machine learning\lab2\a9a.t")

    # 1220
    zeros = np.zeros(16281)
    X_test = np.c_[X_test, zeros]

    # 
    learning_rate = 0.01
    epoch = 300

    NAG_losses = get_NAG(epoch, learning_rate)
    RMSProp_losses = get_RMSProp(epoch, learning_rate)
    AdaDelta_losses = get_AdaDelta(epoch)
    Adam_losses = get_Adam(epoch, learning_rate)

    # loss
    plt.plot(range(epoch), NAG_losses, label="NAG_loss", color='red')
    plt.plot(range(epoch), RMSProp_losses, label="RMSProp_loss", color='blue')
    plt.plot(range(epoch), AdaDelta_losses, label="AdaDelta_loss", color='green')
    plt.plot(range(epoch), Adam_losses, label="Adam_loss", color='yellow')
    plt.xlabel("epochs")
    plt.ylabel('loss')
    plt.legend()
    plt.show()
```
